## CONNECTING TO STORAGE ACCOUNT

```python
storage_account_name = "capitaledgedatalake"
container_name = "capitaledgecontainer"
storage_account_key = dbutils.secrets.get(scope="capitaledgedatalakekey", key="capitaledgedatalakekey")

# configure databricks to access the storage account
spark.conf.set(f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",storage_account_key)

data_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/"

print("Authentication configured")



```
